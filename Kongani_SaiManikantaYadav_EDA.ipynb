```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load datasets
customers = pd.read_csv('Customers.csv')
products = pd.read_csv('Products.csv')
transactions = pd.read_csv('Transactions.csv')
```

```python
# Basic Info
print("Customers Info:\n", customers.info())
print("Products Info:\n", products.info())
print("Transactions Info:\n", transactions.info())
```

```python
# Checking for missing values
print("Missing Values:\n")
print("Customers:\n", customers.isnull().sum())
print("Products:\n", products.isnull().sum())
print("Transactions:\n", transactions.isnull().sum())
```

```python
# Checking for duplicates
print("Duplicate Entries:\n")
print("Customers:", customers.duplicated().sum())
print("Products:", products.duplicated().sum())
print("Transactions:", transactions.duplicated().sum())
```

```python
# Summary statistics
print("Customers Summary:\n", customers.describe(include='all'))
print("Products Summary:\n", products.describe(include='all'))
print("Transactions Summary:\n", transactions.describe())
```

```python
# Data visualization
plt.figure(figsize=(10, 6))
sns.countplot(x='Region', data=customers, order=customers['Region'].value_counts().index)
plt.title('Customer Distribution by Region')
plt.xticks(rotation=45)
plt.show()

```

```python


plt.figure(figsize=(10, 6))
sns.barplot(x='Category', y='Price', data=products)
plt.title('Average Price by Product Category')
plt.xticks(rotation=45)
plt.show()
```

```python


plt.figure(figsize=(10, 6))
sns.histplot(transactions['TotalValue'], bins=30, kde=True)
plt.title('Transaction Value Distribution')
plt.show()
```
